```python
# List of numerical columns from your dataset
numerical_columns = ['age in years', 'height', 'weight', 'ap_hi', 'ap_lo']

# Define a function to treat outliers using IQR
def treat_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)  # Cap the outliers
    return df

# Apply the function to treat outliers
Ccardio = treat_outliers_iqr(Ccardio, numerical_columns)

# Visualize the treated data with boxplots
for column in numerical_columns:
    sns.boxplot(x=Ccardio[column], color='green')
    plt.title(f"Boxplot after treating outliers for {column}")
    plt.tight_layout()
    plt.show()
```

**The table appears to show standardized values (z-scores) for the variables `age in years`, `height`, `weight`, `ap_hi`, and `ap_lo`. Here's a quick breakdown:**

1. **Standardization Process**:
   - The values are transformed to have a mean of 0 and a standard deviation of 1.
   - This makes the data comparable across different scales, which is especially useful for machine learning algorithms sensitive to scale (e.g., gradient descent-based models).

2. **Interpreting Values**:
   - Positive values (e.g., `0.252004` for `age in years`) indicate the data point is above the mean for that variable.
   - Negative values (e.g., `-1.076607` for `height`) indicate the data point is below the mean for that variable.
   - Magnitudes (e.g., `1.430669` for `ap_hi`) reflect how far a value is from the mean in terms of standard deviations.

3. **Example Rows**:
   - **Row 0**: 
     - `age in years` (-0.487534): Below the mean.
     - `height` (0.459209): Above the mean.
   - **Row 3**:
     - `ap_hi` (1.430669): Significantly above the mean.
     - `ap_lo` (2.009954): Even further above the mean, indicating a potential outlier.

**Purpose:**
Standardization is typically done before applying algorithms like logistic regression or SVMs to ensure that all variables contribute equally to the model, avoiding dominance by variables with larger scales.


1. **Correlation Matrix**:
   - Calculates the Pearson correlation coefficient for numerical features to measure linear relationships.
   - Values range from `-1` (strong negative) to `1` (strong positive), with `0` indicating no correlation.

2. **StandardScaler**:
   - Standardizes features using \( Z = \frac{X - \mu}{\sigma} \), where \( X \) is the feature value, \( \mu \) is the mean, and \( \sigma \) is the standard deviation.
   - Ensures all features are on the same scale, crucial for machine learning models sensitive to magnitude differences.

3. **Feature Transformation**:
   - `scaler.fit_transform()` computes \(\mu\) and \(\sigma\), then applies standardization to the numerical columns.
   - Standardized features have a mean of 0 and a standard deviation of 1.

4. **DataFrame Update**:
   - The transformed values replace the original numerical columns in the DataFrame, preparing the dataset for analysis.

This process ensures uniform scaling and readiness for machine learning.

**10.Modelling**

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Assuming 'Ccardio' is your cleaned and standardized dataset
X = Ccardio.drop('cardio', axis=1)  # Features
y = Ccardio['cardio']               # Target
 
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Support Vector Machine': SVC(probability=True, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5)
}

# Train, predict, and evaluate each model
results = []
for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    y_pred = model.predict(X_test)  # Predict on test data
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Append results
    results.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    })

# Create a DataFrame to compare results
results_df = pd.DataFrame(results)

# Print evaluation results
print("Evaluation Metrics for Each Model:")
print(results_df)

# Save the best-performing model
best_model_name = results_df.sort_values(by='F1 Score', ascending=False).iloc[0]['Model']
best_model = models[best_model_name]
print(f"\nBest Model: {best_model_name}")

# Save the best model
import joblib
joblib.dump(best_model, f"{best_model_name.replace(' ', '_').lower()}_model.pkl")
print(f"Best model saved as '{best_model_name.replace(' ', '_').lower()}_model.pkl'")

```

**Insights:**
XGBoost emerges as the best model, striking a balance between Precision (0.748989), Recall (0.680388), and F1 Score (0.713042), making it ideal for tasks requiring balanced metrics.
Support Vector Machine achieves the highest Accuracy (0.733333) but slightly lower Recall, indicating it may miss some positive cases compared to XGBoost.
Other models like Logistic Regression and Random Forest show slightly lower performance.

**Hyperparameter Tuning and Evaluation**

```python
#Using  GridSearchCV

from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Define parameter grid
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 150],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [0.5, 1.0, 2.0]
}

# Initialize the model
xgb_model = XGBClassifier(eval_metric='logloss', random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=3, scoring='f1', verbose=1, n_jobs=-1)

# Fit the model
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best F1 Score:", grid_search.best_score_)
```

**GridSearchCV Summary:**
**Fitting Details:**

**The grid search tested 243 combinations of hyperparameters.**
**It performed 3-fold cross-validation for each combination, resulting in 729 total fits.**
**Best Parameters:**

**'learning_rate': 0.2: A relatively higher learning rate, indicating faster convergence.**

**'max_depth': 3: A shallow tree depth to prevent overfitting.**

**'n_estimators': 150: A reasonable number of boosting rounds.**

**'reg_alpha': 0.5: L1 regularization to reduce overfitting.**

**'reg_lambda': 1.0: Default L2 regularization for further control of model complexity.**

**Best F1 Score:**

**0.7219: The harmonic mean of precision and recall, indicating the model performed well on the validation folds, especially if your dataset has some class imbalance.**


```python
# Train the model with best parameters
best_params = grid_search.best_params_
xgb_tuned = XGBClassifier(
    eval_metric='logloss',
    random_state=42,
    **best_params
)
xgb_tuned.fit(X_train, y_train)

# Evaluate on test data
y_pred = xgb_tuned.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Save the model
import joblib
joblib.dump(xgb_tuned, 'xgboost_tuned_model.pkl')
print("Tuned model saved as 'xgboost_tuned_model.pkl")
```
