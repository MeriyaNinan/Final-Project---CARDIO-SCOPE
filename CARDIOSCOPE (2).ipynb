The choice of **50 and 250** for filtering the height and blood pressure (ap_hi and ap_lo) values is based on the following reasoning:

**Height:**

A typical human height generally ranges between **50** cm (for infants) and around **250** cm (for exceptionally tall individuals, though rare).
50 cm is a reasonable lower bound since it's unlikely for adults to have a height below this threshold.
250 cm is a very high value for an adult, but it's used as the upper bound to remove any extreme or unrealistic outliers in the dataset.

**Blood Pressure (ap_hi and ap_lo):**

Systolic pressure (ap_hi) and diastolic pressure (ap_lo) for a healthy individual usually **range from 90-120** mmHg for systolic **and 60-80** mmHg for diastolic. Anything outside the range of **50-250** mmHg would be considered extremely unrealistic.
50 mmHg is a very low value for both systolic and diastolic pressures, and 250 mmHg is very high, possibly indicating measurement errors or incorrect data entry.
In essence, these thresholds are chosen to remove unrealistic or implausible values based on domain knowledge of human physiology. The goal is to keep only reasonable values while excluding outliers that could distort the analysis.

```python
# List of numerical columns from your dataset
numerical_columns = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']

# Define a function to treat outliers using IQR
def treat_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)  # Cap the outliers
    return df

# Apply the function to treat outliers
Ccardio = treat_outliers_iqr(Ccardio, numerical_columns)

# Visualize the treated data with boxplots
for column in numerical_columns:
    sns.boxplot(x=Ccardio[column], color='green')
    plt.title(f"Boxplot after treating outliers for {column}")
    plt.tight_layout()
    plt.show()
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Assuming 'Ccardio' is your cleaned and standardized dataset
X = Ccardio.drop('cardio', axis=1)  # Features
y = Ccardio['cardio']               # Target
 
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Support Vector Machine': SVC(probability=True, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5)
}

# Train, predict, and evaluate each model
results = []
for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    y_pred = model.predict(X_test)  # Predict on test data
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Append results
    results.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    })

# Create a DataFrame to compare results
results_df = pd.DataFrame(results)

# Print evaluation results
print("Evaluation Metrics for Each Model:")
print(results_df)

# Save the best-performing model
best_model_name = results_df.sort_values(by='F1 Score', ascending=False).iloc[0]['Model']
best_model = models[best_model_name]
print(f"\nBest Model: {best_model_name}")

# Save the best model
import joblib
joblib.dump(best_model, f"{best_model_name.replace(' ', '_').lower()}_model.pkl")
print(f"Best model saved as '{best_model_name.replace(' ', '_').lower()}_model.pkl'")

```

**Hyperparameter Tuning and Evaluation**

```python
#Using GridSearchCV

from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

# Define parameter grid
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [50, 100, 150],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [0.5, 1.0, 2.0]
}

# Initialize the model
xgb_model = XGBClassifier(eval_metric='logloss', random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='f1', verbose=1, n_jobs=-1)

# Fit the model
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best F1 Score:", grid_search.best_score_)
```

```python
# Train the model with best parameters
best_params = grid_search.best_params_
xgb_tuned = XGBClassifier(
    eval_metric='logloss',
    random_state=42,
    **best_params
)
xgb_tuned.fit(X_train, y_train)

# Evaluate on test data
y_pred = xgb_tuned.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Save the model
import joblib
joblib.dump(xgb_tuned, 'xgboost_tuned_model.pkl')
print("Tuned model saved as 'xgboost_tuned_model.pkl'")
```
